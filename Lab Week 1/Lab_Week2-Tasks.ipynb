```python
# TASK: Read the given data, i.e., ‘nai_data.txt’ in text format into a required DataFrame.
# Make the “IMAGE” column be the index of the returned DataFrame.
# Display the returned DataFrame (name it as “nai_data”) in your programme.

nai_data = pd.read_table('/Users/tom/Documents/Data Visulisation and Representation/Lab Week 1/nai_type.txt',index_col='IMAGE')
```

```python
# TASK: Read the given data, i.e., ‘dtype.txt’ in text format into a required DataFrame.
# Make the “IMAGE” column be the index of the returned DataFrame.
# Display the returned DataFrame (name it as “dtype”) in your programme.

d_type = pd.read_table('/Users/tom/Documents/Data Visulisation and Representation/Lab Week 1/dtype.txt',index_col='IMAGE')
```

```python
# TASK: Combine data by making the data contained in “dtype” as a new column of “nai_data”.

nai_data = nai_data.join(d_type,on='IMAGE')
```

```python
# TASK: Extract data from above table so that the new table contains the “ghost” type of images only.
results = nai_data.loc[nai_data['TYPE']=='ghost']
results
```

```python
# TASK: Based on above table, calculate the mean of each image (i.e., each row) and
# construct a new DataFrame that includes the results.
mean_only = results.mean(axis=1,numeric_only=True)
mean_only
```
